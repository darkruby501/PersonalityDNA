```python
fig,(ax1,ax1a) = plt.subplots(1,2,figsize=(12,5),sharey=True)

x1 = responses.notnull().sum(1)
x1a = 100*responses.notnull().sum(1)/696
sns.distplot(x1,ax=ax1,kde=False)
sns.distplot(x1a,ax=ax1a,kde=False)
ax1.set_xlabel('Number of Questions Answered')
ax1a.set_xlabel('Percentage of Questions Answered (%)')
ax1.set_ylabel('Number of Users')
fig.suptitle('Hist: Number of Users for Each Response Rate',fontsize=20)

```

```python
fig,ax2 = plt.subplots(1,1,figsize=(8,5))
x2 = responses.notnull().sum(0)/responses.shape[0]
sns.distplot(x2*100,bins=10,ax=ax2,kde=False,color='g')
ax2.set_ylabel('Number of Questions')
ax2.set_xlabel('Percengage of Users Answering')
ax2.set_title('How many responses did each question get?')
```

Apologies, the above plot is not intuitive. What it shows is the bulk of questions (400+/696) were each answered by around 10% of the respondents. Only a dozen or so questions were answered by a quarter of the participants. The plot below illustrates these points in a clearer manner.

```python
x2 = responses.notnull().sum(0)/responses.shape[0]
sns.barplot(x=list(x2.index),y=x2)
plt.title('Bar Plot: Response Rate for Each Question')
plt.xlabel('Questions, 1-676')
plt.ylabel('Response Rate (Proportion of Respondents)')
```

There is one column for each question. The majority of questions were answered by 10%. A few got a few more. Overall, responses were distributed equally between questions.

```python
k = 300
samples_k = np.random.randint(responses.shape[0],size=k)
B = responses.notnull().iloc[samples_k].values.astype(int)

fig = plt.figure(figsize=(15,15))  # figure size in inches
plt.imshow(B,cmap=plt.cm.binary,extent=[0,676,0,k])
plt.xlabel('Question #')
plt.ylabel('Respondent #')
plt.title('Visualisation of Questions Answered By k Random Respondents')
```

Each row is a respondent, each dot in their row indicates they answered a question in the correspdonding column. Response are distributed evenly. Any two given respondents will not have much overlap.

```python
BB = np.dot(B,B.T)
np.fill_diagonal(BB,0)
fig = plt.figure(figsize=(10,10))  # figure size in inches
plt.imshow(BB,cmap=plt.cm.autumn_r)
plt.colorbar()
plt.title('Colormap of Number of Questions In Common')
```

Each pixel indicates the number of questions two respondents (row and column) have both answered. In the random sample of users, most have no more than a dozen out of 696 questions in common. Some users answered ~300 and therefore have questions in common with many.

```python
(responses.notnull().sum(1)>299).sum()
```

```python
responses_many = responses[responses.notnull().sum(1)>299]
responses_many.fillna(0,inplace=True)
```

```python
CC = np.dot(responses_many,responses_many.T)
np.fill_diagonal(CC,0)
fig = plt.figure(figsize=(10,10))  # figure size in inches
plt.imshow(CC,cmap=plt.cm.inferno_r)
plt.colorbar()
```

```python
responses_many.replace(0,np.nan,inplace=True)
num_responses = []
for i in range(0,responses_many.shape[0]): 
    num_responses.append(responses_many.iloc[i].notnull().sum())
```

```python
from scipy.spatial.distance import euclidean
dist = euclidean
```

```python
n_many = responses_many.shape[0]

EE = np.zeros((n_many,n_many))

if responses_many.isnull().sum().sum()>0:
    responses_many.fillna(0,inplace=True)

for i in range(0,n_many):
    for j in range(0,n_many):
        EE[i,j] = dist(responses_many.iloc[i],responses_many.iloc[j])
```

```python
EE/EE.max()
```

```python
FF = np.fill_diagonal(EE,EE.max())
fig = plt.figure(figsize=(10,10))  # figure size in inches
plt.imshow(EE/EE.max(),cmap=plt.cm.inferno_r)
plt.colorbar()
```

```python

CC = np.dot(responses_many.fillna(0),responses_many.fillna(0).T)

#Let's try normalising by number of questions answered by minimum person.
for i in range(0,responses_many.shape[0]):
    for j in range(0,responses_many.shape[0]):
        CC[i,j] = CC[i,j]/min(num_responses[i],num_responses[j])

        
        
np.fill_diagonal(CC,0)
fig = plt.figure(figsize=(10,10))  # figure size in inches
plt.imshow(CC,cmap=plt.cm.inferno_r)
plt.colorbar()
```
